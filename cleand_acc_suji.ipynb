```python
from collections import Counter
import re

# emp_title의 고유값들을 가져옵니다. (이미 소문자, 공백 제거된 상태라고 가정)
# processed_unique_titles_in_other 에 'Other'에 속한 원본 고유값들이 들어있다고 가정
# 또는 df_cat['emp_title'].unique() 로 현재 그룹화된 emp_title의 고유값들을 사용
sample_emp_titles = list(df_cat['emp_title'].unique()) # 현재 df_cat에 그룹화된 emp_title 사용 예시

# 만약 group_emp_title로 아직 안 묶인 상태의 emp_title_clean을 쓴다면
# sample_emp_titles = acc_df_cleaned['emp_title'].astype(str).str.lower().str.strip().unique()


# 텍스트 전처리 (불필요한 문자 제거 등)
def clean_text(text):
    text = str(text)
    text = re.sub(r'[^a-zA-Z\s]', '', text) # 알파벳과 공백만 남기기
    text = text.lower()
    return text

cleaned_titles = [clean_text(title) for title in sample_emp_titles]

# 단어 빈도 분석
all_words = []
for title in cleaned_titles:
    all_words.extend(title.split())

word_counts = Counter(all_words)

print("--- 가장 흔한 단어 (Top 50) ---")
for word, count in word_counts.most_common(50):
    print(f"{word}: {count}")

# 특정 n-gram (예: 2-gram) 분석
from nltk.util import ngrams
# import nltk # 필요 시 nltk.download('punkt')
# from nltk.tokenize import word_tokenize

# all_bigrams = []
# for title in cleaned_titles:
#     tokens = word_tokenize(title)
#     all_bigrams.extend(list(ngrams(tokens, 2)))

# bigram_counts = Counter(all_bigrams)
# print("\n--- 가장 흔한 2-gram (Top 20) ---")
# for bigram, count in bigram_counts.most_common(20):
#     print(f"{bigram}: {count}")
```

```python
from collections import Counter # 단어 빈도 계산을 위해 필요합니다.
import re # 텍스트 클리닝을 위해 필요합니다.
import os # 파일 저장 경로 확인을 위해 필요합니다.


# --- 0. acc_df_cleaned 데이터 프레임이 준비되어 있고, 'emp_title' 컬럼이 존재한다고 가정합니다 ---
# acc_df_cleaned['emp_title']에는 NaN이 있을 수 있으므로 처리합니다.

# --- 1. emp_title의 고유값들을 가져와 전처리합니다 (소문자, 공백 제거) ---
# 먼저 NaN 값을 제거한 후, 소문자화 및 공백 제거를 수행하여 고유값들을 추출합니다.
# .astype(str)은 NaN을 'nan' 문자열로 바꾸는 것을 방지합니다.
# filter(pd.notna)를 사용하여 NaN을 먼저 제거합니다.
all_original_emp_titles = acc_df_cleaned['emp_title'].dropna().astype(str).str.lower().str.strip()
unique_processed_emp_titles = all_original_emp_titles.unique()


# --- 2. 텍스트 클리닝 함수 정의 ---
# 단어 빈도 분석 시 방해되는 특수문자 등을 제거합니다.
def clean_text_for_word_analysis(text):
    text = str(text) # 혹시 None 등이 있다면 문자열로 변환
    # 알파벳, 숫자, 공백 외의 모든 문자 제거 (단어 추출을 위함)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # 여러 개의 공백을 하나의 공백으로 대체
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# --- 3. 각 직업명을 단어로 분리하고 빈도를 계산 ---
all_words = []
for title in unique_processed_emp_titles:
    cleaned_title = clean_text_for_word_analysis(title)
    if cleaned_title: # 빈 문자열이 아닌 경우에만 단어 추가
        all_words.extend(cleaned_title.split())

word_counts = Counter(all_words)


# --- 4. 단어 빈도수를 Pandas DataFrame으로 변환 ---
# 컬럼 이름은 'word'와 'count'로 지정합니다.
df_word_counts = pd.DataFrame(word_counts.items(), columns=['word', 'count'])

# 빈도수(count) 기준으로 내림차순 정렬 (선택 사항)
df_word_counts = df_word_counts.sort_values(by='count', ascending=False).reset_index(drop=True)


# --- 5. 결과 확인 (상위 50개) ---
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)

print("--- 'emp_title' 단어 빈도 분석 결과 (상위 50개) ---")
print(df_word_counts.head(50))


# --- 6. 결과를 CSV 파일로 저장 ---
output_csv_filename = 'emp_title_word_frequencies.csv'

try:
    df_word_counts.to_csv(
        output_csv_filename,
        index=False,         # DataFrame의 인덱스를 CSV에 포함하지 않음
        encoding='utf-8-sig' # 한글 깨짐 방지를 위해 'utf-8-sig' 사용
    )
    print(f"\n단어 빈도 분석 결과가 '{output_csv_filename}' 파일에 성공적으로 저장되었습니다.")
    print(f"파일 저장 위치: {os.getcwd()}")

except Exception as e:
    print(f"CSV 파일 저장 중 오류 발생: {e}")

# --- (선택 사항) Pandas 출력 옵션 초기화 ---
# pd.reset_option('display.max_rows')
# pd.reset_option('display.max_colwidth')
# pd.reset_option('display.width')
```

```python
!pip install scikit-learn

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# --- 0. 필요한 데이터 준비 ---
# acc_df_target과 df_cat은 이미 준비되어 있고, 다음과 같은 상태라고 가정합니다.
# - acc_df_target: 'loan_status_binary' 컬럼이 있는 DataFrame (원본 데이터 필터링됨)
# - df_cat: 'emp_title_professional_level' 컬럼이 생성되고 분류된 DataFrame
#           (df_cat의 인덱스는 acc_df_target의 인덱스와 일치해야 합니다.)

# 로지스틱 회귀에 사용할 데이터를 하나의 DataFrame으로 합칩니다.
# acc_df_target의 해당 인덱스에 df_cat['emp_title_professional_level']을 직접 추가합니다.
if 'emp_title_professional_level' in df_cat.columns:
    combined_df = acc_df_target[['loan_status_binary']].copy()
    combined_df['emp_title_level'] = df_cat['emp_title_professional_level'] # 새로운 이름으로 사용

    # 만약 그룹화 과정에서 NaN이 있는 행이 df_cat에서 제거되었다면,
    # combined_df에 NaN이 생길 수 있습니다. 이 NaN은 다시 제거해야 합니다.
    combined_df.dropna(subset=['emp_title_level'], inplace=True)

    print(f"결합된 DataFrame 크기: {combined_df.shape}")
    print(combined_df.head())
    print(combined_df.info())
else:
    print("df_cat에 'emp_title_professional_level' 컬럼이 없거나, 올바른 DataFrame이 아닐 수 있습니다.")
    print("데이터 준비 단계를 확인해주세요.")
    # 다음 단계를 진행할 수 없습니다.


# --- 1. 독립 변수 (X) 와 종속 변수 (y) 분리 ---
X = combined_df[['emp_title_level']] # 독립 변수: 직업 전문성 레벨
y = combined_df['loan_status_binary'] # 종속 변수: 대출 상태 (0 또는 1)

print(f"\nX 형태: {X.shape}, y 형태: {y.shape}")


# --- 2. 원-핫 인코딩 (범주형 변수 처리) ---
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

X_encoded_array = encoder.fit_transform(X)

# 인코딩된 배열을 DataFrame으로 변환 (컬럼명 유지)
X_encoded = pd.DataFrame(X_encoded_array, columns=encoder.get_feature_names_out(['emp_title_level']), index=X.index)

print(f"원-핫 인코딩 후 X 형태: {X_encoded.shape}")
print("원-핫 인코딩된 X 상위 5개 행:")
print(X_encoded.head())


# --- 3. 훈련 세트와 테스트 세트 분리 ---
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

print(f"\n훈련 세트 X_train 형태: {X_train.shape}, y_train 형태: {y_train.shape}")
print(f"테스트 세트 X_test 형태: {X_test.shape}, y_test 형태: {y_test.shape}")


# --- 4. 로지스틱 회귀 모델 학습 ---
model = LogisticRegression(solver='liblinear', max_iter=200, random_state=42)
model.fit(X_train, y_train)

print("\n로지스틱 회귀 모델 학습 완료.")

```

```python
# --- 5. 모델 평가 ---
y_pred = model.predict(X_test)

print("\n--- 모델 평가 ---")
accuracy = accuracy_score(y_test, y_pred)
print(f"정확도 (Accuracy): {accuracy:.4f}")

print("\n분류 보고서 (Classification Report):")
print(classification_report(y_test, y_pred))

print("\n혼동 행렬 (Confusion Matrix):")
print(confusion_matrix(y_test, y_pred))

# 각 직업 전문성 레벨에 대한 계수 (Coefficients) 확인
print("\n--- 모델 계수 (Coefficients) ---")
coefficients = pd.DataFrame({'Feature': X_encoded.columns, 'Coefficient': model.coef_[0]})
coefficients['Absolute_Coefficient'] = np.abs(coefficients['Coefficient'])
coefficients = coefficients.sort_values(by='Absolute_Coefficient', ascending=False)
print(coefficients) # 직업 전문성 레벨은 4개 뿐이니 전체를 출력

print(f"\n절편 (Intercept): {model.intercept_[0]:.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder # 라벨 인코딩
from sklearn.metrics import classification_report, accuracy_score

# --- 1. transformers 라이브러리 임포트 ---
# pip install transformers torch 또는 pip install transformers tensorflow 로 먼저 설치하세요.
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from torch.utils.data import DataLoader, TensorDataset # 딥러닝 데이터 처리


# --- 0. 데이터 준비 ---
# acc_df_cleaned, df_cat (emp_title_professional_level 포함)이 준비된 상태라고 가정.

# 사용할 입력 특성: 그룹화 전의 emp_title 클린 버전
# emp_title_cleaned 는 df_cat['emp_title'].astype(str).str.lower().str.strip() 후 NaN 제거된 상태
# 이전에 'emp_title_clean' 컬럼을 별도로 만들지 않았다면,
# df_cat에 현재 그룹화된 'emp_title'을 임시로 'emp_title_cleaned_input'으로 사용합니다.
# 하지만 가장 좋은 것은, 그룹화되기 전의 원본 직업명(소문자/공백 제거된)을 사용하는 것입니다.
# 여기서는 df_cat에 emp_title_professional_level과 함께 존재하는 원본 'emp_title' (NaN 없고 소문자 처리됨)을 사용합니다.

# 데이터셋 크기를 줄일 샘플 개수를 지정합니다.
sample_size = 50000 # 예시: 5만 개의 데이터만 사용

# 전체 데이터가 샘플 크기보다 클 경우에만 샘플링 로직 적용
if df_cat.shape[0] > sample_size:
    print(f"데이터셋을 {sample_size}개로 샘플링합니다.")
    
    # 1. 먼저 전체 데이터프레임을 훈련/테스트 세트로 나누는 것처럼 샘플링합니다.
    #    stratify를 사용하여 'emp_title_professional_level'의 비율을 유지합니다.
    #    이때 test_size를 사용하여 원하는 샘플 크기만큼만 '테스트' 세트에 할당되도록 합니다.
    #    예: 전체 데이터가 200만개, sample_size가 5만개이면 test_size = 5만 / 200만 = 0.025
    
    # sample_size가 너무 작아 train_test_split의 오류를 유발할 경우 대비
    if sample_size <= df_cat.shape[0] * 0.01 and sample_size < 5: # 최소 샘플 개수
        print("경고: 샘플 크기가 너무 작아 train_test_split 오류를 유발할 수 있습니다. 최소 50개 권장.")
    
    _, df_cat_sampled = train_test_split(
        df_cat, # 전체 df_cat을 대상으로
        test_size=sample_size / df_cat.shape[0], # 원하는 sample_size 만큼 테스트 세트에 할당
        random_state=42,
        stratify=df_cat['emp_title_professional_level'] # 라벨 비율 유지
    )
    # '_' 변수는 사용하지 않는 train_test_split의 훈련 세트 부분입니다.
    # df_cat_sampled에는 stratify된 샘플 데이터가 들어갑니다.
    print(f"샘플링된 데이터셋 크기: {df_cat_sampled.shape[0]}개")
else:
    df_cat_sampled = df_cat.copy()
    print("전체 데이터셋 크기가 지정된 샘플 크기보다 작아 샘플링하지 않았습니다.")
    print(f"사용될 데이터셋 크기: {df_cat_sampled.shape[0]}개")

# 입력 텍스트 데이터 (이미 소문자화, 공백 제거, NaN 제거된 emp_title)
#texts = df_cat['emp_title'].tolist() # df_cat['emp_title']이 이미 전처리된 직업명이라고 가정
texts = df_cat_sampled['emp_title'].tolist() # df_cat['emp_title']

# 타겟 라벨 (emp_title_professional_level)
# labels_text = df_cat['emp_title_professional_level'].tolist()

labels_text = df_cat_sampled['emp_title_professional_level'].tolist()

# 라벨 인코딩: 문자열 라벨을 숫자로 변환
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels_text) # ['Management', 'Non-Professional', 'Other', 'Professional'] -> [0, 1, 2, 3]

# 클래스 개수 확인
num_labels = len(label_encoder.classes_)
print(f"인코딩된 라벨 클래스: {label_encoder.classes_}")
print(f"총 라벨 개수: {num_labels}")
```

```python
# --- 데이터 분리 ---
# 학습용과 평가용 데이터셋으로 분리
train_texts, val_texts, train_labels, val_labels = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)
print(f"훈련 데이터 개수: {len(train_texts)}, 검증 데이터 개수: {len(val_texts)}")


# --- 2. 토크나이저 및 모델 로드 ---
# 사용할 BERT 모델 지정 (예: 'bert-base-uncased' - 대소문자 구분 없음)
model_name = 'bert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_name)

# 분류를 위한 BERT 모델 로드 (num_labels에 따라 분류 레이어 생성)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

# GPU 사용 가능 여부 확인
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
print(f"\n모델을 {device} 장치로 로드했습니다.")


# --- 3. 데이터 토큰화 및 BERT 입력 형식으로 변환 ---
def encode_texts(tokenizer, texts, max_length=128):
    # max_length는 BERT 모델이 처리할 수 있는 최대 시퀀스 길이 (직업명이 길지 않으므로 128로 충분)
    # padding='max_length': 모든 시퀀스를 max_length에 맞춰 패딩
    # truncation=True: max_length를 초과하는 시퀀스 자르기
    # return_tensors='pt': PyTorch 텐서로 반환 (tf: TensorFlow 텐서) 
    return tokenizer(texts, padding='max_length', truncation=True, max_length=max_length, return_tensors='pt')

# 훈련 데이터 인코딩
train_encodings = encode_texts(tokenizer, train_texts)
train_dataset = TensorDataset(
    train_encodings['input_ids'],
    train_encodings['attention_mask'],
    torch.tensor(train_labels, dtype=torch.long)
)

# 검증 데이터 인코딩
val_encodings = encode_texts(tokenizer, val_texts)
val_dataset = TensorDataset(
    val_encodings['input_ids'],
    val_encodings['attention_mask'],
    torch.tensor(val_labels, dtype=torch.long)
)

# DataLoader 생성 (배치 학습을 위함)
batch_size = 16 # GPU 메모리가 허용하는 범위 내에서 설정. CPU에서는 작게.
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

print(f"\n데이터 로더 생성 완료. 배치 크기: {batch_size}")


# --- 4. (선택적) 모델 학습 루프 (매우 기본적인 예시, 실제 학습에는 Trainer API 권장) ---
# 이 부분은 GPU 없이는 매우 오래 걸릴 수 있습니다.
# 실제 모델 학습을 위해서는 Hugging Face Trainer API 또는 PyTorch Lightning 등 사용을 강력히 권장합니다.

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5) # AdamW는 딥러닝 학습에 흔히 사용되는 최적화 기법
num_epochs = 3 # 예시 에폭 수

print("\n--- 모델 학습 시작 (CPU에서는 매우 느릴 수 있습니다) ---")
model.train() # 모델을 훈련 모드로 설정
for epoch in range(num_epochs):
     total_loss = 0
     for batch in train_loader:
         optimizer.zero_grad() # 이전 기울기 초기화
         input_ids, attention_mask, labels = [b.to(device) for b in batch] # 데이터를 GPU/CPU로 이동
         outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
         loss = outputs.loss
         total_loss += loss.item()
         loss.backward() # 역전파
         optimizer.step() # 가중치 업데이트

     avg_train_loss = total_loss / len(train_loader)
     print(f"에폭 {epoch+1}/{num_epochs}, 평균 훈련 손실: {avg_train_loss:.4f}")

     print("--- 모델 학습 완료 ---")


# --- 5. 모델 평가 (학습된 모델이 있다고 가정) ---
# 여기서는 모델이 이미 학습되었다고 가정하고 검증 데이터로 평가합니다.
# (위의 학습 루프를 주석 처리했다면, 랜덤 가중치로 인한 매우 낮은 성능을 보일 것입니다.)
print("\n--- 모델 평가 시작 ---")
model.eval() # 모델을 평가 모드로 설정 (드롭아웃 등을 비활성화)
val_preds = []
val_true = []

with torch.no_grad(): # 기울기 계산 비활성화 (메모리 절약, 속도 향상)
    for batch in val_loader:
        input_ids, attention_mask, labels = [b.to(device) for b in batch]
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits # 예측 결과 (softmax 전)
        predictions = torch.argmax(logits, dim=-1) # 가장 높은 확률을 가진 클래스 선택

        val_preds.extend(predictions.cpu().numpy()) # GPU -> CPU -> NumPy
        val_true.extend(labels.cpu().numpy())

print("\n분류 보고서 (Classification Report):")
print(classification_report(val_true, val_preds, target_names=label_encoder.classes_))
print(f"정확도 (Accuracy): {accuracy_score(val_true, val_preds):.4f}")


# --- 6. 새로운 직업명 예측 예시 ---
print("\n--- 새로운 직업명 예측 예시 ---")
new_job_titles = ["software developer", "team manager", "truck driver", "senior financial analyst", "student intern", "rn clinical nurse"]
model.eval() # 예측 모드
for job_title in new_job_titles:
    encoded_input = encode_texts(tokenizer, job_title)
    input_ids = encoded_input['input_ids'].to(device)
    attention_mask = encoded_input['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        predicted_label_id = torch.argmax(logits, dim=-1).item()
        predicted_category = label_encoder.inverse_transform([predicted_label_id])[0]

    print(f"직업명: '{job_title}' -> 예측된 분류: '{predicted_category}'")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# imbalanced-learn 라이브러리 임포트 (설치 필요: pip install imbalanced-learn)
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline # 파이프라인을 사용하여 전처리 및 모델 학습을 효율적으로 통합

# --- 0. 필요한 데이터 준비 ---
# combined_df는 이전에 생성된 DataFrame이며,
# emp_title_professional_level을 X, loan_status_binary를 y로 사용합니다.

# ==================== 수정된 부분 시작 ====================
# BERT 학습을 위해 샘플링된 df_cat_sampled를 사용하는 것이 아니라,
# 로지스틱 회귀 분석을 위해 원본 combined_df를 다시 사용합니다.
# 만약 메모리 문제가 없다면 전체 데이터를 사용하는 것이 좋습니다.
# 여기서는 combined_df에서 X_encoded와 y를 가져옵니다.

# 로지스틱 회귀 독립 변수 X_encoded와 종속 변수 y 준비 (이전 코드에서 이미 생성)
# combined_df['emp_title_level']을 X, combined_df['loan_status_binary']를 y로.
# X는 원-핫 인코딩된 상태여야 합니다.

# X = combined_df[['emp_title_level']]
# encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
# X_encoded = pd.DataFrame(encoder.fit_transform(X), columns=encoder.get_feature_names_out(['emp_title_level']), index=X.index)
# y = combined_df['loan_status_binary']

# 이 변수들이 정의되어 있지 않다면, 위 주석 처리된 부분을 다시 실행하여 생성해 주세요.
# ==================== 수정된 부분 끝 ====================

print(f"모델 학습 데이터 X_encoded 형태: {X_encoded.shape}, y 형태: {y.shape}")

# --- 1. 훈련 세트와 테스트 세트 분리 (데이터 불균형 처리 전) ---
# 데이터 불균형 처리는 훈련 세트에만 적용해야 합니다!
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

print(f"\n불균형 처리 전 훈련 세트 (y_train) 클래스 분포:")
print(y_train.value_counts(normalize=True))
print(f"불균형 처리 전 테스트 세트 (y_test) 클래스 분포:")
print(y_test.value_counts(normalize=True))


# --- 2. SMOTE 적용을 위한 파이프라인 생성 및 모델 학습 ---
# SMOTE는 오직 훈련 세트에만 적용합니다.
# 훈련 시에 SMOTE로 오버샘플링하고, 로지스틱 회귀 모델을 학습하는 파이프라인을 만듭니다.
# 'smote': SMOTE 변환기
# 'classifier': LogisticRegression 모델

model_pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)), # 소수 클래스 오버샘플링
    ('classifier', LogisticRegression(solver='liblinear', max_iter=200, random_state=42))
])

print("\nSMOTE와 로지스틱 회귀 모델 학습 시작...")
model_pipeline.fit(X_train, y_train) # 파이프라인 학습


# --- 3. 모델 평가 (원본 테스트 세트 사용) ---
# 불균형 처리가 안 된 원본 테스트 세트(X_test, y_test)를 사용하여 모델을 평가합니다.
y_pred = model_pipeline.predict(X_test)

print("\n--- SMOTE 적용 후 모델 평가 ---")
accuracy = accuracy_score(y_test, y_pred)
print(f"정확도 (Accuracy): {accuracy:.4f}")

print("\n분류 보고서 (Classification Report):")
print(classification_report(y_test, y_pred))

print("\n혼동 행렬 (Confusion Matrix):")
print(confusion_matrix(y_test, y_pred))

# 학습된 로지스틱 회귀 모델의 계수 확인
# 파이프라인의 'classifier' 단계에 접근합니다.
coefficients = pd.DataFrame({'Feature': X_encoded.columns, 'Coefficient': model_pipeline.named_steps['classifier'].coef_[0]})
coefficients['Absolute_Coefficient'] = np.abs(coefficients['Coefficient'])
coefficients = coefficients.sort_values(by='Absolute_Coefficient', ascending=False)
print("\n--- SMOTE 적용 후 모델 계수 (Coefficients) ---")
print(coefficients)

print(f"\n절편 (Intercept): {model_pipeline.named_steps['classifier'].intercept_[0]:.4f}")
```
