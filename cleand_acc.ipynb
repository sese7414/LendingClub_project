```python
# 'rn'이 포함되면 무조건 'nurse'로 먼저 처리 (최우선 순위)
rn_condition = grouped_emp_title.str.contains('rn', na=False)
grouped_emp_title.loc[rn_condition] = 'nurse'
```

```python
for keyword in job_keywords:
    condition = grouped_emp_title.str.contains(keyword, na=False)
    grouped_emp_title.loc[condition] = keyword # 조건에 맞는 값을 키워드로 업데이트

grouped_emp_title[~grouped_emp_title.isin(job_keywords + ['nurse'])] = 'Other'

# 원래 컬럼에 덮어 쓰기
df_cat['emp_title'] = grouped_emp_title.astype('category')
```

```python
# 'id' 컬럼에 'Total amount funded' 문자열이 포함된 행을 찾습니다.
# na=False는 NaN 값을 False로 처리하여 포함하지 않게 합니다.
# case=False는 대소문자 구분 없이 찾습니다.
metadata_row = acc_df_cleaned['id'].astype(str).str.contains('Total amount funded', na=False, case=False)

print(f"식별된 메타데이터 행의 개수: {metadata_row.sum()}개")

# --- 2. 해당 행들을 DataFrame에서 제거 ---
# 이 불필요한 행들을 제외한 새로운 DataFrame을 생성합니다.
# '~' (틸다)는 조건의 반대(True인 것을 False로, False인 것을 True로)를 의미합니다.
acc_df_cleaned_after_str_contains = acc_df_cleaned[~metadata_row].copy()

print(f"원본 acc_df_cleaned 행 수: {acc_df_cleaned.shape[0]}개")
print(f"`str.contains()` 후 acc_df_cleaned_after_str_contains 행 수: {acc_df_cleaned_after_str_contains.shape[0]}개")

# 제거 후의 결측치 상태 확인 (제거되었음을 확인)
target_nan_cols_to_check = ['term', 'grade', 'home_ownership', 'verification_status', 'purpose']
print("\n--- `str.contains()` 후 주요 컬럼의 결측치 개수 재확인 ---")
print(acc_df_cleaned_after_str_contains[target_nan_cols_to_check].isnull().sum())
```
