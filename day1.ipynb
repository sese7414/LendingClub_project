특정 키워드로 직업군 묶기

```python
# 통합할 대표 키워드 리스트
keywords_to_group = [
    'manager', 'nurse', 'driver', 'supervisor', 'sales', 
    'officer', 'director', 'engineer', 'technician', 'president',
    'teacher', 'owner','assistant','operator','service']

# 키워드 포함 여부에 따라 분류, 없으면 원래 이름 유지
def group_job_title(title):
    if pd.isna(title):
        return np.nan
    for keyword in keywords_to_group:
        if keyword in title:
            return keyword
    return title  # 키워드 포함 안 되면 원래 값 그대로 유지

# 적용: emp_title_clean 기준으로 통합 컬럼 생성
selected_df['emp_title_grouped'] = selected_df['emp_title_clean'].apply(group_job_title)
# 'emp_title_grouped'은 전처리한 결과 칼럼에서 특정 키워드끼리 포함한 결과를 담은 칼럼!!!!!!!!!!!!!

# 결과 확인
print(selected_df['emp_title_grouped'].value_counts().head(30))
```

```python
unique_job_count = selected_df['emp_title_grouped'].nunique()
print(f"'emp_title_grouped'의 고유 직업명 개수: {unique_job_count}")
```

```python
# rn를 통일시키는 코드 2번 확인용 
keywords_to_group = [
    'manager', 'nurse', 'driver', 'supervisor', 'sales', 
    'officer', 'director', 'engineer', 'technician', 'president',
    'teacher', 'owner', 'assistant', 'operator', 'service'
]

def group_job_title(title):
    if pd.isna(title):
        return np.nan
    
    # 먼저 rn을 nurse로 통일 처리 rn이 약어라고 함.
    if 'rn' in title:
        return 'nurse'
    
    for keyword in keywords_to_group:
        if keyword in title:
            return keyword
    return title  # 키워드 포함 안 되면 원래 값 그대로 유지

selected_df['emp_title_grouped'] = selected_df['emp_title_clean'].apply(group_job_title)

print(selected_df['emp_title_grouped'].value_counts().head(50))
```

```python
unique_job_count = selected_df['emp_title_grouped'].nunique()
print(f"'emp_title_grouped'의 고유 직업명 개수: {unique_job_count}")
```

수지님이 주신 코드로 확인해보는 직업수 총 정리판

```python
job_keywords = [
    'engineer', 'manager', 'driver', 'teacher', 'analyst',
    'specialist', 'assistant', 'technician', 'sales', 'clerk',
    'developer', 'nurse', 'attorney', 'accountant', 'consultant',
    'coordinator', 'representative', 'supervisor', 'admin', 'operator',
    'director', 'medical', 'finance', 'architect', 'researcher',
    'advisor', 'designer', 'mechanic', 'electrician', 'welder',
    'marketing', 'production', 'training', 'officer', 'business',
    'owner', 'partner', 'president', 'vp',  # 'vp' = 부사장
    'executive', 'intern', 'student', 'retired', 'other'  # 'other'는 특별 처리
]

# 'emp_title_clean'을 기준으로 시작
grouped_emp_title = selected_df['emp_title_clean'].copy()
grouped_emp_title = grouped_emp_title.astype('category')
grouped_emp_title = grouped_emp_title.cat.add_categories(['Other'])  # 'Other' 카테고리 추가

# 'rn'이 포함되면 nurse로 우선 처리
rn_condition = grouped_emp_title.str.contains('rn', case=False, na=False)
grouped_emp_title.loc[rn_condition] = 'nurse'

# job_keywords에 따라 분류
for keyword in job_keywords:
    condition = grouped_emp_title.str.contains(keyword, case=False, na=False)
    grouped_emp_title.loc[condition] = keyword

# 지정된 키워드 외 값은 'Other'로 처리
grouped_emp_title[~grouped_emp_title.isin(job_keywords + ['nurse'])] = 'Other'

# 새 컬럼으로 저장
selected_df['emp_title_grouped'] = grouped_emp_title.astype('category')

# --- 결과 확인 ---
print(f"--- 'emp_title_grouped' 컬럼 그룹화 결과 ---")
print(f"그룹화 후 고유값 개수: {selected_df['emp_title_grouped'].nunique()}개")

print("\n상위 20개 직업 빈도:")
print(selected_df['emp_title_grouped'].value_counts().head(20))

print("\n하위 10개 직업 빈도:")
print(selected_df['emp_title_grouped'].value_counts().tail(10))

# nurse와 Other 카테고리 확인
print("\n'nurse' 카테고리 빈도:")
if 'nurse' in selected_df['emp_title_grouped'].cat.categories:
    print(selected_df['emp_title_grouped'].value_counts()['nurse'])
else:
    print("'nurse' 카테고리 없음.")

print("\n'Other' 카테고리 빈도:")
if 'Other' in selected_df['emp_title_grouped'].cat.categories:
    print(selected_df['emp_title_grouped'].value_counts()['Other'])
else:
    print("'Other' 카테고리 없음.")
```
